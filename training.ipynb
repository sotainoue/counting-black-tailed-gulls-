```python
#folders = glob.glob('/home/inoue/Desktop/kabushima/count/ortho/*')
folders = glob.glob('/media/inoue/KABU3/orthomosaic/24*')  
#folders = ['/media/inoue/KABU1/orthomosaic/240417_5','/media/inoue/KABU1/orthomosaic/240417_6']
#folders = ['/home/inoue/Desktop/kabushima/count/ortho/230511_1']

for folder in folders:
    print(folder)
    images = glob.glob(folder +'/*.tif')
    
    
    tmp3  = {'image_id': ['NN'],
            'bbox': ['NN'],
            'score': ['NN'],
            'category_id': ['NN']}
    tmp3 = pd.DataFrame(tmp3)
    
    for image in images:
        
        result = get_sliced_prediction(image,
                                       detection_model,
                                       slice_height=640,
                                       slice_width=640,
                                       overlap_height_ratio=0.2,
                                       overlap_width_ratio=0.2,
                                       postprocess_match_threshold = 0.3)
        
        if len(result.to_coco_predictions()) > 0:
            tmp = pd.DataFrame(result.to_coco_predictions())
            tmp = tmp.drop('segmentation',axis=1)
            tmp = tmp.drop('category_name',axis=1)
            tmp = tmp.drop('iscrowd',axis=1)
            tmp = tmp.drop('area',axis=1)
        
            #os.path.split(image)[1]
            tmp['image_id'] = os.path.split(image)[1]
    
            tmp3 = pd.concat([tmp3,tmp],ignore_index=True)
    
        tmp3.to_csv('/media/inoue/KABU3/orthomosaic/prediction/' + os.path.split(image)[1] + '.csv')
```
